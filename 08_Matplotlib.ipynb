# Visualización de gráficos con Matplotlib


Matplotlib es un paquete de Python diseñado para la visualización de gráficos de forma ágil y sencilla utilizando comandos básicos. 

## Importando los paquetes necesarios

```python
import matplotlib.pyplot as plt
import numpy as np   # Paquete de cálculo científico y análisis de datos
import pandas as pd  # Paquete de ánalisis de datos
```

## Carga de archivos

Vamos a trabajar con el archivo Ventas creado cuando se presentó Panda.

```python
Datos=pd.read_excel('VentasXls.xlsx', sheet_name='VentasXls') ## carga de archivos
Datos
list(Datos)
```

## Gráfica de línea

Un gráfico de línea es utilizado para visualizar la información de cada variable para un individuo en específico. Es muy útil para representar series de tiempo.\

A continuación se realiza un gráfico de la venta realizada en cada uno de los registros. Observe que en el eje x se coloca el número de registro y en el eje y la respectiva venta.

```python
plt.plot(Datos['Unnamed: 0'],Datos["Venta"],label="Ventas") ## Gráfica
plt.xlabel('Registro') #Nombre eje x
plt.ylabel('Pesos')    #Nombre eje y 
plt.title("Ventas")    #Titulo del gráfico
plt.legend()           #Leyenda
plt.show()             #Visualización
```

## Gráfica de descuento y comisión

Varias variables en un mismo gráfico.

```python
plt.plot(Datos["Descuento"],label="Descuento",marker="o",color="b") ## Descuento
plt.plot(Datos["Comision"],label="Comisión",marker=">",color="r") ## Comisión

plt.xlabel('Registro') #Nombre eje x
plt.ylabel('Pesos')    #Nombre eje y 
plt.title("Descuento y Comisión")    #Titulo del gráfico
plt.legend()           #Leyenda
plt.show()             #Visualización
```

## Varios Gráficos

Es posible realizar varios gráficos en una misma ventana utilizando el comando $plt.subplot$

```python
plt.subplot(131)       # Divide el dispositivo gráfico
plt.plot(Datos["Venta"],label="Ventas")
plt.xlabel('Registro') #Nombre eje x
plt.ylabel('Pesos')    #Nombre eje y 
plt.title("Ventas")    #Titulo del gráfico
plt.subplot(132)
plt.plot(Datos["Descuento"],label="Descuento")
plt.xlabel('Registro') #Nombre eje x
plt.title("Descuento")    #Titulo del gráfico
plt.subplot(133)
plt.plot(Datos["Comision"],label="Comision")
plt.xlabel('Registro') #Nombre eje x
plt.title("Comision")    #Titulo del gráfico
```

## Diagramas de dispersión

Un diagrama de dispersión 2D es un gráfico de puntos que representa el valor que toma cada individuo en dos variables distintas. Se grafica el de las variables Venta y Descuento.

```python
plt.scatter(Datos["Venta"],Datos["Descuento"])

plt.xlabel('Ventas')                #Nombre eje x
plt.ylabel('Descuento')             #Nombre eje y 
plt.title("Ventas vs Descuento")    #Titulo del gráfico
```

También, es posible cambiar el tipo de marca que se quiere visualizar con el comando $marker$, el tamaño con $s$ y el color con $c$. A continuación, se crea un vector aleatorio $x$ para que cada marca quede con un color distinto.

```python
x = np.random.rand(len(Datos))        # Selección de colores
plt.scatter(Datos["Comision"],Datos["Descuento"],c=x, marker="+",s=80)
plt.xlabel('Comisión')                #Nombre eje x
plt.ylabel('Descuento')               #Nombre eje y 
plt.title("Comisión vs Descuento")    #Titulo del gráfico
```

Ahora, un gráfico con todas las posibles diagramas de dispersión.

```python
plt.subplot(331)
plt.scatter(Datos["Venta"],Datos["Venta"],s=5, marker=">",color="r")
plt.ylabel("Ventas")    #Nombre eje y 
plt.title("Ventas")    #Titulo del gráfico
plt.subplot(334)
plt.scatter(Datos["Venta"],Datos["Descuento"],s=5, marker="+",color="y")
plt.ylabel("Descuento")    #Nombre eje y 
plt.subplot(337)
plt.scatter(Datos["Venta"],Datos["Comision"],s=5, marker="D",color="b")
plt.ylabel("Comision")    #Nombre eje y 


plt.subplot(332)
plt.scatter(Datos["Descuento"],Datos["Venta"],s=25, marker=">",color="r")
plt.title("Descuento")    #Titulo del gráfico
plt.subplot(335)
plt.scatter(Datos["Descuento"],Datos["Descuento"],s=25, marker="+",color="y")
plt.subplot(338)
plt.scatter(Datos["Descuento"],Datos["Comision"],s=25, marker="D",color="b")


plt.subplot(333)
plt.scatter(Datos["Comision"],Datos["Venta"],s=50, marker=">",color="r")
plt.title("Ventas")    #Titulo del gráfico
plt.subplot(336)
plt.scatter(Datos["Comision"],Datos["Descuento"],s=50, marker="+",color="y")
plt.subplot(339)
plt.scatter(Datos["Comision"],Datos["Comision"],s=50, marker="D",color="b")


```

## Histograma

Un histrograma es un diagrama de frecuencias para una variable cuantitativa continua.

### Histograma de la variable ventas

```python
plt.hist(Datos["Venta"],color="r",bins=30) 
#bins es la cantidad de intervalos

plt.ylabel("Frecuencia")
plt.xlabel("Pesos")
plt.title("Histograma Ventas")
```

### Diagrama de cajas

O diagrama de bigotes, son gráficos que representan la mediana, los cuartiles y puntos atípicos.

```python
plt.boxplot(Datos["Venta"]) 

plt.xlabel("Ventas")

```

### Diagrama de torta para una variable cualitativa

Los diagramas de tortas representan en un gráfico circular el porcentaje de individuos que poseen determinada característica.

```python
Valores=Datos["Vendedor"].value_counts() / len(Datos["Vendedor"])
Valores  ### Valores a graficar


plt.pie(Valores,labels=["María","Juana","Jose"],autopct="%1.1f%%")
plt.title("Cantidad de ventas realizadas por cada vendedor")
```
